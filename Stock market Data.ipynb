```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

### Fetching Live Stock Data

```python
#stock ticker symbol 
ticker = "AAPL"
```

```python
stock_data = yf.Ticker(ticker)
stock_data
```

```python
data = stock_data.history(period="1d", interval="1m")
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

##### Open: The price at which the stock opened during the given time interval.
##### High: The highest price the stock reached during the interval.
##### Low: The lowest price the stock reached during the interval.
##### Close: The price at which the stock closed at the end of the interval.
##### Volume: The number of shares traded during the interval.
##### Dividends: The dividend paid out, if any, during the time interval.
##### Stock Splits: Any stock splits during the interval (if any).

##### returnOnEquity: A profitability ratio that measures a company's ability to generate profits from its shareholders' equity.
##### freeCashflow: The amount of cash a company generates after accounting for capital expenditures.
##### operatingCashflow: Cash generated from the company's core business operations.
##### earningsGrowth: The percentage increase in a company's earnings over time.
##### revenueGrowth: The rate at which a company's revenue increases over a given period.
##### grossMargins: The percentage of total sales revenue that the company retains after incurring the direct costs associated with producing goods and services.
##### ebitdaMargins: EBITDA (Earnings Before Interest, Taxes, Depreciation, and Amortization) margin, which shows how much of a company’s earnings come from operations.
##### operatingMargins: The percentage of revenue left after paying for variable costs of production.
##### financialCurrency: The currency in which the financials are reported (USD in this case).
##### trailingPegRatio: Price/earnings to growth ratio that accounts for future growth.
